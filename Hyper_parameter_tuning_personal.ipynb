## Hyper-Parameter Tuning for Random Forest

```python
# Hyper-Parameter Search Grid Using 10-Fold CV and Test
print(' -- Random Forest --')
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import cross_val_score 
from sklearn.metrics import roc_curve, auc


n_estimators = [50,100,200]
criterion=['gini', 'entropy']
bootstrap= [True, False]
max_depth=[10, 20,30]



train_results_mean = []
train_results_std = []

train_results_auc=[]
test_results_auc=[]

test_accu=[]
test_precision=[]
features=[]
for estimator in n_estimators:
    for cr in criterion:
        for boots in bootstrap:
            for max_d in max_depth:
                
                rfc = RandomForestClassifier(n_estimators=estimator,
                                             criterion=cr,
                                             bootstrap=boots,
                                             max_depth=max_d, 
                                             class_weight={0:1-y_train.mean(), 1:y_train.mean()},
                                             random_state=0,
                                             n_jobs=-1)
    
    
                all_accuracies = cross_val_score(estimator=rfc, X=X_train, y=y_train, cv=10)
                train_results_mean.append(all_accuracies.mean())
                train_results_std.append(all_accuracies.std())
    
                rfc.fit(X_train, y_train)
    
                train_pred = rfc.predict(X_train)
    
                precision,recall,F1,accuracy,confusion,roc_auc=scores.scores(y_train,train_pred)
                train_results_auc.append(roc_auc)
    
                y_pred = rfc.predict(X_test)
    
                precision,recall,F1,accuracy,confusion,roc_auc=scores.scores(y_test,y_pred)
                test_results_auc.append(roc_auc)
                test_accu.append(accuracy)
                test_precision.append(precision)
                features.append([estimator,cr,boots,max_d])
                #print('Average CV mean accuracy',all_accuracies.mean())
                #print('Test Set accuracy',accuracy)
                #print(estimator,cr,boots,max_d)
        
    


    

    
    
```

```python
df1= pd.DataFrame({'train_results_mean':train_results_mean})
df2= pd.DataFrame({'train_results_std':train_results_std})
df3= pd.DataFrame({'train_results_auc':train_results_auc})
df4= pd.DataFrame({'test_results_auc':test_results_auc})
df5= pd.DataFrame({'test_accuracy':test_accu})
df6= pd.DataFrame( {'test_precision':test_precision})
df7= pd.DataFrame( {'features':features})

df_results=pd.concat([df1, df2,df3,df4,df5,df6,df7],axis=1)
df_results[df_results['test_accuracy']==df_results['test_accuracy'].max()]
```

```python
df1= pd.DataFrame({'train_results_mean':train_results_mean})
df2= pd.DataFrame({'train_results_std':train_results_std})
df3= pd.DataFrame({'train_results_auc':train_results_auc})
df4= pd.DataFrame({'test_results_auc':test_results_auc})
df5= pd.DataFrame({'test_accuracy':test_accu})
df6= pd.DataFrame( {'test_precision':test_precision})
df7= pd.DataFrame( {'features':features})

df_results=pd.concat([df1, df2,df3,df4,df5,df6,df7],axis=1)
df_results[df_results['test_accuracy']==df_results['test_accuracy'].max()]
```
