## Now do a for loop for each element in order to replace it for a numpy array and not a string and make it the "y"

```python
y=[]
for x in stab_vector:
    #print(x)
    a=np.fromstring(x[1:-1],sep=',').astype(int)
    y.append(a)
y=np.array(y)    
```
